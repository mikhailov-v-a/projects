```python
import bs4
import requests
import re
```

```python
n = 4 # к-во страниц
list_cat = []
for i in range (1, n+1):
    url_cat = 'https://www.bafus.ru/sq/djvo/page:' + str(i) + '/f1:72788/'
    page_cat = requests.get(url_cat)
    soup_cat = bs4.BeautifulSoup(page_cat.text, 'html.parser')
    all_cat = soup_cat.findAll('a', class_='product-link sub-catalog')
    pattern = re.compile(r'href="(.*)/"')
    list_cat = list_cat + (re.findall(pattern, str(all_cat)))
list_cat = list(set(list_cat))
```

```python
list_cat
```

```python
target = []
for j in list_cat:
    url_target = j
    page_target = requests.get(url_target)
    soup_target = bs4.BeautifulSoup(page_target.text, 'html.parser')
    all_target = soup_target.findAll('a', class_='product-link')
    pattern = re.compile(r'href="(.*)/"')
    target = target + (re.findall(pattern, str(all_target)))
print(target)
```
