```python
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
```

```python
file_names = [1,2,3]
tags = ['OPEN_DATE', 'RELATIONSHIP', 'FINAL_PMT_DATE', 
        'TYPE', 'PMT_STRING_84M', 'STATUS', 'OUTSTANDING', 'INF_CONFIRM_DATE', 
        'CREDIT_LIMIT', 'DELQ_BALANCE', 'IS_OWN', 'CREDIT_COST_RATE', 'UUID']
df = pd.DataFrame(columns=tags)
val_dict = {}


for name in file_names:
    file = open(str(name) + '.xml', 'r')
    xml_file = file.read()
    soup = BeautifulSoup(xml_file, 'xml')
    
    group_id = soup.find('GROUP_ID').text
    create_datetime = soup.find('CREATE_DATETIME').text
    
    test_nan = []
    
    for tag in tags:
        val_list = []
        
        for i in soup.findAll(tag):
            val_list.append(i.text)
        if test_nan == val_list:
            val_list.append(np.nan)
            
        val_dict[tag] = val_list
    
    val_dict['GROUP_ID'] = [group_id] * len(val_list)
    val_dict['DATE'] = [create_datetime] * len(val_list)
    
    new_df = pd.DataFrame(val_dict)
    df = df.append(new_df, ignore_index=True)
    file.close
    
df.loc[:, 'DATE'] = df.DATE.apply(lambda x: str(x)[0:2]+'.'+str(x)[2:4]+'.'+str(x)[4:8])
df.head()
```

```python
df.to_excel('part_3.xlsx', sheet_name='Часть 3', index=False)
```
